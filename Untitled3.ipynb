```python
import zipfile
import os

# Zip dosyasının yolu (Colab'da genellikle '/content' dizininde olur)
zip_path = '/content/images.zip'

# Çıkarılacak klasörün yolu
extract_path = '/content/images'

# Eğer çıkarılacak klasör yoksa oluştur
if not os.path.exists(extract_path):
    os.makedirs(extract_path)

# Zip dosyasını aç ve içeriğini çıkar
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print(f"'{zip_path}' dosyası '{extract_path}' klasörüne başarıyla çıkarıldı.")

# Çıkarılan dosyaları listele
extracted_files = os.listdir(extract_path)
print("\nÇıkarılan dosyalar:")
for file in extracted_files:
    print(file)

# Created/Modified files during execution:
print(f"{extract_path}/{file}" for file in extracted_files)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
from tensorflow.keras.optimizers import Adam
import os

# CSV dosyasını yükle
df = pd.read_csv('train.csv')

# Kategori sütununu string'e çevir
df['category'] = df['category'].astype(str)

# Görüntü dosyalarının yolunu kontrol et
image_dir = '/content/images/images'  # CSV dosyasının bulunduğu dizin
if not os.path.exists(image_dir):
  raise FileNotFoundError(f"'{image_dir}' klasörü bulunamadı.")

# Görüntü dosyalarının varlığını kontrol et
image_files = os.listdir(image_dir)
valid_images = [f for f in df['image'] if f in image_files]
if len(valid_images) == 0:
  raise FileNotFoundError("Hiçbir görüntü dosyası bulunamadı.")

print(f"Toplam {len(valid_images)} geçerli görüntü dosyası bulundu.")

# Görüntüleri yüklemek için bir veri artırma nesnesi oluştur
datagen = ImageDataGenerator(rescale=1./255, validation_split=0.2)

# Eğitim ve doğrulama setleri
train_generator = datagen.flow_from_dataframe(
  dataframe=df,
  directory=image_dir,
  x_col='image',
  y_col='category',
  target_size=(150, 150),
  batch_size=32,
  class_mode='categorical',
  subset='training'
)

validation_generator = datagen.flow_from_dataframe(
  dataframe=df,
  directory=image_dir,
  x_col='image',
  y_col='category',
  target_size=(150, 150),
  batch_size=32,
  class_mode='categorical',
  subset='validation'
)

# Modeli oluştur
model = Sequential([
  Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
  MaxPooling2D(pool_size=(2, 2)),
  Conv2D(64, (3, 3), activation='relu'),
  MaxPooling2D(pool_size=(2, 2)),
  Flatten(),
  Dense(128, activation='relu'),
  Dropout(0.5),
  Dense(5, activation='softmax')  # 5 kategori
])

# Modeli derle
model.compile(optimizer=Adam(), loss='categorical_crossentropy', metrics=['accuracy'])

# Modeli eğit
if train_generator.samples > 0:
  history = model.fit(train_generator, validation_data=validation_generator, epochs=10)

  # Eğitim sürecini görselleştir
  plt.plot(history.history['accuracy'], label='Eğitim Doğruluğu')
  plt.plot(history.history['val_accuracy'], label='Doğrulama Doğruluğu')
  plt.title('Model Doğruluğu')
  plt.ylabel('Doğruluk')
  plt.xlabel('Epoch')
  plt.legend()
  plt.show()
else:
  print("Eğitim için yeterli görüntü bulunamadı.")
```

```python
# Eğitim ve doğrulama doğruluğunu görselleştir
plt.plot(history.history['accuracy'], label='Eğitim Doğruluğu')
plt.plot(history.history['val_accuracy'], label='Doğrulama Doğruluğu')
plt.title('Model Doğruluğu')
plt.ylabel('Doğruluk')
plt.xlabel('Epoch')
plt.legend()
plt.show()

# Eğitim ve doğrulama kaybını görselleştir
plt.plot(history.history['loss'], label='Eğitim Kaybı')
plt.plot(history.history['val_loss'], label='Doğrulama Kaybı')
plt.title('Model Kaybı')
plt.ylabel('Kayıp')
plt.xlabel('Epoch')
plt.legend()
plt.show()
```

```python
import os
import zipfile
import numpy as np
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import tempfile

try:
  # Modeli yükle
  model = load_model('gemi_tespiti_model.h5')
  print("Model başarıyla yüklendi.")

  # ZIP dosyasının yolu
  zip_path = '/content/test_images.zip'  # ZIP dosyasının tam yolu
  print(f"ZIP dosyası: {zip_path}")

  # Geçici bir dizin oluştur
  with tempfile.TemporaryDirectory() as temp_dir:
      print(f"Geçici dizin oluşturuldu: {temp_dir}")

      # ZIP dosyasını geçici dizine çıkar
      with zipfile.ZipFile(zip_path, 'r') as zip_ref:
          zip_ref.extractall(temp_dir)
      print("ZIP dosyası açıldı ve içeriği geçici dizine çıkarıldı.")

      # test_images klasörünün yolunu oluştur
      test_images_dir = os.path.join(temp_dir, 'test_images')

      # test_images klasöründeki tüm görselleri al
      image_files = [f for f in os.listdir(test_images_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
      print(f"Toplam {len(image_files)} görüntü dosyası bulundu.")

      if len(image_files) == 0:
          print("Uyarı: Hiç görüntü dosyası bulunamadı!")

      # Her bir görsel için tahmin yap
      for img_file in image_files:
          img_path = os.path.join(test_images_dir, img_file)
          img = image.load_img(img_path, target_size=(150, 150))
          img_array = image.img_to_array(img) / 255.0  # Normalizasyon
          img_array = np.expand_dims(img_array, axis=0)  # Boyutunu genişlet

          # Tahmin yap
          predictions = model.predict(img_array)
          predicted_class = np.argmax(predictions, axis=1)

          print(f'{img_file} için tahmin edilen kategori: {predicted_class[0]}')

  print("İşlem tamamlandı.")

except Exception as e:
  print(f"Bir hata oluştu: {str(e)}")
```
